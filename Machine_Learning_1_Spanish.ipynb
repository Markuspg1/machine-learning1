```python
def load_csv(fn):
    
    import pandas as pd
    
    print("Cargando datos desde ", fn, end=" ...")
    df = pd.read_csv(fn)
    print("Completado.")
    print("Filas y Columnas: ", df.shape) #muestra el numero de filas y columnas
    print("Columnas\n", df.columns)
    print()
    print("df.columns.values.tolist()\n", df.columns.values.tolist()) #convierte las columnas en una lista
    
    return df

print("Funcion Definida")
```

```python
#https://stackoverflow.com/questions/24458645/label-encoding-across-multiple-columns-in-scikit-learn
from sklearn.preprocessing import LabelEncoder
class MultiColumnLabelEncoder:
    def __init__(self,columns = None):
        self.columns = columns # array of column names to encode

    def fit(self,X,y=None):
        return self # not relevant here

    def transform(self,X):
        '''
        Transforms columns of X specified in self.columns using
        LabelEncoder(). If no columns specified, transforms all
        columns in X.
        '''
        output = X.copy()
        if self.columns is not None:
            for col in self.columns:
                output[col] = LabelEncoder().fit_transform(output[col])
        else:
            for colname,col in output.iteritems():
                output[colname] = LabelEncoder().fit_transform(col)
        return output

    def fit_transform(self,X,y=None):
        return self.fit(X,y).transform(X)
```

# Arbol de Clasificacion

```python
import pandas as pd
import numpy as np
```

```python
# Agregamos la columna clasificacion de precios
conditions = [
    (df['SalePrice'] <= 100000),
    (df['SalePrice'] > 100000) & (df['SalePrice'] <= 500000),
    (df['SalePrice'] > 500000)
    ]

# La lista de grupos
values = ['Group 1', 'Group 2', 'Group 3']

# Creamos la nueva columna con los valores segun la condicion
df['Group'] = np.select(conditions, values)

# display updated DataFrame
df.head()
```

```python
dfClass = df[['OverallQual','GrLivArea','GarageCars','GarageArea','TotalBsmtSF','1stFlrSF','FullBath','TotRmsAbvGrd','YearBuilt','YearRemodAdd','Group']]
```

```python
X = dfClass.iloc[:,:-1]
X.head(5)
```

```python
y = dfClass.iloc[:,-1]
y.head(5)
```

```python
from sklearn.model_selection import train_test_split as tts
x_train, x_test, y_train, y_test = tts(X,y, test_size=0.2)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
modelo = DecisionTreeClassifier(random_state=0)
arbol_c =modelo.fit(x_train,y_train)
```

```python
y_pred= modelo.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix
C=confusion_matrix(y_pred, y_test)
print(C)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(8,6))
tree.plot_tree(arbol_c, feature_names=list(X.columns.values),
              class_names = list(y.values), filled=True)
plt.show()
```

```python
#Del ejemplo https://colab.research.google.com/drive/1UkA6t4Y9W-f-bVDzLxJp5aYUzxP2J984#scrollTo=4n-EwTzh2N11
# Aqu vamos a implementar k-fold cross validation para k=5, obtendremos 5 accuracy
from sklearn.model_selection import cross_val_score
scores = cross_val_score(modelo, X, y,
                         scoring="accuracy",
                         cv=4, n_jobs=-1)  #cv es k
print(scores)
print(np.mean(scores))
```
